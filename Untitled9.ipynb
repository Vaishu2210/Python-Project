```python
#we are building neural networks by building those value objects
class Value:
    def __init__(self,data,_children=(),_op='',label=' '):
        self.data=data
        self.grad=0.0 #means no effect so at initialization every value doesnt affect the output (changing this variable is not changing the loss function)
        self._backward=lambda:None  #for leaf node
        self._prev=set(_children)
        self._op=_op
        self.label=label

    
 
    def __repr__(self):     #python will internally use this wrapper function to retur nthis string
       return f"Value(data={self.data}, grad={self.grad})"

#python doesnt know how to add two value objects so tell it
    def __add__(self,other):  #we have to use this double underscore to define this operators in python for thes objects
        other =other if isinstance(other,Value) else Value(other)  # if other is a instance of value its fine otherwise we will wrap it up in one of these object
        out=Value(self.data + other.data,(self,other),'+')  #its just wrapping the data so now it has become the addition of two float numbers

        def _backward():
         self.grad+=out.grad  #it overwritres the gradient as we are doing addition so self and other are same object thats why we
         other.grad+=out.grad
        out._backward=_backward # we  are not calling function we are storing them
        return out
        
    def __radd__(self, other): # other + self
        return self + other


    def __pow__(self,other):
        assert isinstance(other,(int,float)), "only supporting int /float powers for now"
        out=Value(self.data**other,(self,), f'**{other}') #other is that power #ask this one

        def _backward():
         self.grad+=other* self.data**(other-1)*out.grad #this self.data has now become a floar or int its not a value object we are accessing now
        out._backward = _backward
        return out
    
    def __rmul__(self,other):  # whwn we have simple __mul__-> x.__mul__(y) but rmul will take this x and y and reverse it y__mul__(x)
        return self*other     #python doesnt know how to do 2*a so it will check if a multiply 2
        
    def __mul__(self,other): 
        other =other if isinstance(other,Value) else Value(other)
        out=Value(self.data * other.data,(self,other),'*') 

        def _backward():
         self.grad+=other.data*out.grad
         other.grad+=self.data*out.grad
        out._backward=_backward                                        
        return out

      # a+(-b)-> a-b
    def __neg__(self):  #-self
        return self*-1  #self.__neg__()

    def __sub__(self, other):  # self-other
        return self+ (-other)   # this is going to neg then it is coming out as -other and we are adding it to self
        

    def __truediv__(self,other):
        return self* other**-1 #after this we have to implement pow function b raise to power-1 



   
    def tanh(self):
        x=self.data
        t=(math.exp(2*x)-1)/(math.exp(2*x)+1)
        out=Value(t,(self,),'tanh')

        def _backward():
         self.grad=(1-t**2)*out.grad  #out is chained throught 1-t2 int o self
        out._backward=_backward
        return out

    def exp(self):
        x=self.data
        out=Value(math.exp(x),(self,),'exp')

        def _backward():
         self.grad+=out.data*out.grad #d/dx of ex is ex
        out._backward=_backward

        return out
        
#we rid of _backward manually  so we have to know all the full dependencies everythingbefore we back  everthing propagate before we continue back propagation
#so this can be achieved by topological sort in which nodes-edges go through left to rights
#it will add o when all of its children are processed
    
    def backward(self):
     topo = []
     visited = set()
     def build_topo(v):
        if v not in visited:
            visited.add(v)
        for child in v._prev:
            build_topo(child)
        topo.append(v)
     build_topo(self)
    
     self.grad=1.0
     for node in reversed(topo):
      node._backward()

a=Value(2.0,label='a')
b=Value(-3.0,label='b')  #we created the value objects
c=Value(8.0,label='c')
e=a*b; e.label='e'
d=e +c;d.label='d'#((a.__mlu__(b)).__add__(c)
f=Value(-2.0,label='f')
L=d*f;L.label='l'
d  #wehere is connective tissue? to know what vlayes are giving out other valyues
        
```

```python
#to visulaize the expression graph mainly forward pass
from graphviz import Digraph
def trace(root):              #trace enumerates all nodes and edges
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):   #for all this nodes then we create a special node objects using dot.node an then we also create edges using dot.edge 
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) #, node_attr={'rankdir': 'TB'})
    nodes,edges= trace(root)
    for n in nodes:
        uid=str(id(n))
        dot.node(name=uid, label = "{%s | data %.4f | grad %.4f}" % (n.label,n.data,n.grad), shape='record')
        if n._op:        #here it is not actual nodes this are just for lookking these are no actual nodes for the gtaoh
            dot.node(name=uid + n._op, label=n._op)
            dot.edge(uid+ n._op,uid)
    
    for n1, n2 in edges:
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)
    
    return dot
```

```python

class Neuron:

    def __init__(self, nin):  # nin is no of inputs for eg 3
        self.w=[Value(random.uniform(-1,1)) for _ in range(nin)] #these are also inputs to this neural nets 
        self.b=Value(random.uniform(-1,1))

    def __call__(self,x):
          #w*x+b
          act=sum((wi*xi for wi, xi in zip(self.w, x)),self.b)
          out=act.tanh()
          return out

    def parameters(self):#gather up all parameters so we can operate on them simultaneouly and we nudge them a tiny little based on gradient info
        return self.w +[self.b]

class Layer:
    def __init__(self,nin,nout): #we initialize independent neurons wiht this given dimensionality
        self.neurons=[Neuron(nin) for _ in range(nout)]  # layer is just a list of neurons where how many neurons we have is taken as input argment and how many neurons do you wnat is number of output in this layer 

    def __call__(self,x):
        outs=[n(x) for n  in self.neurons] # when we call it we just independently evslute them
        return outs[0] if len(outs) ==1 else outs #it returns list i want single element

    def parameters(self):
        return [p for neuron in self.neurons for p in neuron.parameters()]
      


class MLP:
    def __init__(self,nin,nouts): #insteadd of taking single nout which is no of neurons in a single layer,we are going to take list of nouts and this list defines the size of all the layers that we want in our mlp
        sz=[nin]+nouts
        self.layers=[Layer(sz[i],sz[i+1]) for i in range(len(nouts))] # iterate over consecutive pairs of this sizes & create layer objects forthem

    def __call__(self,x):
        for layer in self.layers:
            x=layer(x)
        return x

    def parameters(self):
        return [p for layer in self.layers for p in layer.parameters()]
        
      

#in mlp what is layer of neurons?
#this one layer have number of neurons which are not connected to eath other but fully connected to inputs 

#through micrigrad we can be=ackpropagate all the weights in mlp
```
